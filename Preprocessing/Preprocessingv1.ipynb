```python
def preprocessing(df,target=True):
    from imblearn.under_sampling import RandomUnderSampler
    from sklearn.preprocessing import StandardScaler
    from pickle import dump, load
    
    if target :
        X = df.drop('Survived',axis = 1)
        y = df[['Survived']]
        sampler = RandomUnderSampler()
        X, y = sampler.fit_resample(X,y)
        y['Survived'] = y['Survived'].map({"No":0,"Yes":1})
    else:
        X = df.copy()
           
    X = pd.get_dummies(X,columns=['Sex','Embarked'],drop_first=True)
    features = X.keys().to_list()
    
    if target:
        features = features.append('Survived')

        scaler = StandardScaler()
        scaler.fit(X[['Age','Fare']])
        X[['Age','Fare']] =scaler.transform(X[['Age','Fare']])
        
        # save the scaler
        dump(scaler, open('..\models\preprocessing\standardScalerv1.pkl', 'wb'))
    else:
        #load
        scaler = load(open('..\models\preprocessing\standardScalerv1.pkl', 'rb'))
        X[['Age','Fare']] =scaler.transform(X[['Age','Fare']])
    
    X = pd.DataFrame(X,columns= features)   

    if target:
        df = pd.concat([X,y],axis =1 )
        return df
    else: 
        return X
```

```python
df = pd.read_csv("../Datasets/train_clean.csv",index_col='PassengerId')
```

```python
df.head()
```

```python
df = preprocessing(df)
```

```python
df.to_csv('../Datasets/preprov1_train.csv',index_label='PassengerId')
```
