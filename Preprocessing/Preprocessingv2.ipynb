```python
import pandas as pd
import numpy as np
```

```python
def preprocessingv2(df,target=True):
#     from imblearn.under_sampling import RandomUnderSampler
    from sklearn.decomposition import PCA
    from sklearn.preprocessing import StandardScaler
    from pickle import dump, load
    
    if target :
        X = df.drop('Survived',axis = 1)
        y = df[['Survived']]
#         sampler = RandomUnderSampler()
#         X, y = sampler.fit_resample(X,y)
#         y['Survived'] = y['Survived'].map({"No":0,"Yes":1})
    else:
        X = df.copy()
           
#     X = pd.get_dummies(X,columns=['Sex','Embarked'],drop_first=True)
    features = X.keys().to_list()
    
    if target:
        features = features.append('Survived')
        
        pca = PCA(0.90)
        pca.fit(X)
        X = pca.transform(X)

#         scaler = StandardScaler()
#         scaler.fit(X[['Age','Fare']])
#         X[['Age','Fare']] =scaler.transform(X[['Age','Fare']])

        
        # save the scaler
        dump(pca, open('..\models\preprocessing\pCAv2.pkl', 'wb'))
    else:
        #load
        pca = load(open('..\models\preprocessing\pCAv2.pkl', 'rb'))
        X = pca.transform(X)
#         X[['Age','Fare']] =scaler.transform(X[['Age','Fare']])
    
    X = pd.DataFrame(X)   

    if target:
        df = pd.concat([X,y],axis =1 )
        df = df.rename(columns=lambda x: f' feature {str(x)}' if x != 'Survived' else x)
        return df
    else: 
        X = X.rename(columns=lambda x: f' feature {str(x)}' if x != 'Survived' else x)
        return X
```

```python
df = pd.read_csv('../Datasets/preprov1_train.csv',index_col='PassengerId')
```

```python
df = preprocessingv2(df)
```
